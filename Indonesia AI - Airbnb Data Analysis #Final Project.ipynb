### 5. Get rid of the missing value:

#### a. Drop columns which has missing value that cannot be filled with any logical value (like name, host_id, etc)

```python
data.shape
```

```python
data.isnull().sum()
```

```python
num5 = data.dropna(axis=0, subset=['name','host_name'])
```

```python
num5.isnull().sum()
```

```python
num5.shape
```

#### b. Fill the missing value in last_review with the most common date

```python
num5['last_review'].mode()[0]
```

```python
num5['last_review'] = num5['last_review'].fillna(num5['last_review'].mode()[0])
```

```python
num5.isnull().sum()
```

```python
num5
```

#### c. Fill the missing value in reviews_per_month with the mean

```python
num5['reviews_per_month'] = num5['reviews_per_month'].fillna(np.mean(num5['reviews_per_month']))
```

```python
num5.isnull().sum()
```

```python
num5.shape
```

```python
num5
```

### 6. Make category based on the price: Cheap, Medium, and Expensive

```python
new_df = num5
```

```python
new_df['price'].describe()
```

```python
conditions = [
    (new_df['price'] < 69),
    (new_df['price'] > 69) & (new_df['price'] < 175),
    (new_df['price'] > 175)
]

values = ['Cheap', 'Medium', 'Expensive']

new_df['class'] = np.select(conditions, values)
new_df
```

### 8. Visualize the price to compare the price of available rooms

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')
```

```python
data = new_df
```

```python
#data = new_df.loc[:,['price','availability_365']]
#data
```

```python
data1 = data.groupby(['room_type'])['price'].mean().reset_index()
y = data1['price']
x = data1['room_type']

plt.bar(x, y)
plt.xlabel('Room-type',fontsize='15')
plt.ylabel('Mean Price', fontsize='15')
plt.title('Prices based on room-types',fontsize='15')
plt.show()
```

```python
price = data.loc[:,['availability_365','price']].set_index('availability_365')
price_stats = data['price'].describe().reset_index()
price_counts = price.price.value_counts().reset_index()
price_counts.rename(columns={'index':'price','price':'count'},inplace=True)
fig2,ax = plt.subplots(figsize=(12,8))

plt.hist(price_counts['price'],bins=30,color='k',edgecolor='salmon')
ax.set_xticks(range(0,10000,500))
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
plt.xlabel('Price',fontsize='15')
plt.ylabel('Rentals', fontsize='15')
plt.xlim((-0.5,10000))
plt.title('Price-Rental Distribution',fontsize='15')
plt.show()
```
